```python
def poly_fx(x, y):
    #f = x**2*y + y**2 + 8
    #fx = 0.9999*(1.0*x**1.0 + 0.155*y**1.0 + 0.500001) 
    fx = 1.0*(1.0*x**1.0 + 0.0*y**1.0 + 0.100001) 
    #fx = 1.0/(1.0*x**1.0 + 0.0*y**1.0 + 1.100001) 
    return fx
```

```python
def poly_fy(x, y):
    #f = x**2*y + y**2 + 8
    #fy = 0.9999*(0.144*x**1.0 + 1.0*y**1.0 + 1.70001) 
    fy = 1.0*(0.0*x**1.0 + 1.0*y**1.0 + 0.00001)
    #fy = 1.0/(0.0*x**1.0 + 1.0*y**1.0 + 1.00001)
    return fy
```
